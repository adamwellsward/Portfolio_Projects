# Prediction Visualization

```python
import pandas as pd 
import re
import matplotlib.pyplot as plt
```

```python
# helper function for finding specific character lines
def create_individual_mask(substring, full_string_column):
    # Use apply to check each element in the column
    return full_string_column.apply(lambda x: bool(re.search(re.escape(substring), x)))

# initialize main characters
main_chars = ["Aang", "Sokka", "Katara", "Zuko", "Iroh", "Toph", "Azula"]
```

```python
# import labeled dataset from Kaggle notebook
atla = pd.read_csv("atla_with_predictions.csv").drop(columns=["Unnamed: 0.1", "Unnamed: 0"])

# plot by character
for char in main_chars:
    df = atla[create_individual_mask(char, atla["Character"])]
    df.plot(kind="hist", y=["Label"])
    plt.xticks(ticks=[0, 1, 2, 3], labels=["positive", "negative", "uncertain", "litigious"])
    plt.title(f"{char}")
    plt.show()
```

#### This pattern is to be expected. I hoped to be able to show change over time in frequency of negative/positive lines by each character to show character arcs. Unfortunately, this model was trained on data from Twitter (now X); thus, most of the lines were predicted to be negative or litigious (argumentative). I initially trained using this model because of the conversational nature of Twitter which would match the dialogue from the TV show. In the future I would train a simpler model on positive/negative sentiment only and try to visualize changes over time.
