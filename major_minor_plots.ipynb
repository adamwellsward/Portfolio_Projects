```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
major_df = pd.read_csv('major_results.csv')
major_df.replace([np.nan], np.inf, inplace=True)
minor_df = pd.read_csv('minor_results.csv')
minor_df.replace([np.nan], np.inf, inplace=True)
```

```python
def plot_accuracies(df, data_key='major', val_data_key='all', marker='$m$'):
    # if model_key == "all":
    #     model_col = 'all_key_model_acc'
    # else:
    #     model_col = 'single_key_model_acc'
    # s_k_model_ac = df.loc[df['val_data_key']==val_data_key, ['lambda',model_col]]
    s_k_model_ac = df.loc[df['val_data_key']==val_data_key, ['lambda', 'single_key_model_acc']]
    all_k_model_ac = df.loc[df['val_data_key']==val_data_key, ['lambda', 'all_key_model_acc']]

    plt.plot([str(label) for label in s_k_model_ac['lambda']], s_k_model_ac['single_key_model_acc'], marker=marker, markersize=10, label=f"{data_key} model on {val_data_key} key validation")
    plt.plot([str(label) for label in all_k_model_ac['lambda']], all_k_model_ac['all_key_model_acc'], marker='.', markersize=10, label=f"all key model on {val_data_key} key validation")
    plt.ylabel(f'Val Accuracy')
    plt.xlabel(f'$\lambda$')
plt.rcParams['figure.figsize']=(10, 5)
plt.title('Validation Accuracies Using Key Splits')
plot_accuracies(minor_df, 'minor', 'minor', '$m$')
plot_accuracies(minor_df, 'minor','all', '$m$')
# plt.legend()
# plt.show()
plot_accuracies(major_df, 'major', 'major', '+')
plot_accuracies(major_df, 'major', 'all', '+')
plt.legend()
plt.show()
```
