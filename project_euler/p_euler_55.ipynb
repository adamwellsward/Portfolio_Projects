## Project Euler Solution #55

```python
# functions
def next_iter(n):
    """Returns the next iteration of the 'reverse and add' process. """
    # reverse the number
    rev = int(str(n)[::-1])
    # add it
    return n + rev

def is_palindrome(n):
    num = str(n)
    return num == num[::-1]
```

```python
# iterate through all numbers under 10_000 to find Lychrel numbers
N = 10_000
lychrels = []
for i in range(N):
    num = i
    pal = False
    for _ in range(50):
        new = next_iter(num)
        if is_palindrome(new):
            pal = True
            break
        num = new
    if not pal:
        lychrels.append(i)
print(f"There are {len(lychrels)} Lychrel numbers under {N}.")    
```
