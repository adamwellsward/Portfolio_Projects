## Project Euler Solution #58

```python
import numpy as np
from important_functions import is_prime
from time import time
prime = np.vectorize(is_prime)
```

```python
# function adapted from Problem 28
def spiral_diags(n):
    """Compute the diagonals of a number spiral of size n x n, where n is an odd number."""
    if n % 2 == 0:
        raise ValueError(f"n must be an odd number. {n} is even.")
    
    diags = [1]             # 1 is the center
    num_rings = (n + 1) / 2 # the number of rings in an n x n spiral

    ring = 1
    val = 1
    while ring < num_rings:
        for _ in range(4):  # for each of the 4 corners
            val += 2*ring   # add 2*ring to the last value
            diags.append(val)
        ring += 1
    
    return np.array(diags)
```

```python
# test with side length 7
N = 7
diags = spiral_diags(N)
print(f"The prime ratio for the spiral of side length {N} is {sum(prime(diags))} to {len(diags)}, or {100*sum(prime(diags)) / float(len(diags)):.1f}%")
```

```python
# run the search
start = time()
N = 100_001
full_diags = spiral_diags(N)
print(f"{time() - start:.3f} seconds to load diags")
prime_diags = prime(full_diags)
n = 7
found = False
print(f"Searching... (started at {time() - start:.3f} seconds)")
while not found:
    n += 2
    num_diags = n*2 - 1
    diags = full_diags[:num_diags]
    p_diags = prime_diags[:num_diags]
    if sum(p_diags) / float(num_diags) < .1:
        found = True

print(f"The side length is {n}.")
```
