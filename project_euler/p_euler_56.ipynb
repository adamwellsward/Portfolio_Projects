## Project Euler Solution #56

```python
# run through the numbers
max_dig_sum = 0
result = None
for a in range(2, 100):
    for b in range(1, 100):
        dig_sum = sum([int(d) for d in str(a**b)])
        if dig_sum > max_dig_sum:
            max_dig_sum = dig_sum
            result = (a, b, a**b)

print(f"The max digital sum is {max_dig_sum} and comes from a = {result[0]}, b = {result[1]}, and a ** b = {result[2]}.")
```
