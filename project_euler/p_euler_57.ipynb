## Project Euler Solution #57

```python
def n_iterations(n):
    """Return the num and denom of n iterations of the sequence. """
    first = (3, 2)
    iters = [first]
    if n < 1:
        raise ValueError("n must be at least 1")
    if n > 1:
        for i in range(n-1):
            num, denom = iters[i]
            new_denom = num + denom
            new_num = denom + new_denom
            iters.append((new_num, new_denom))
    return iters
```

```python
# run through all of the terms of the sequence up to N and count the digits
N = 1000
count = 0
for num, denom in n_iterations(N):
    if len(str(num)) > len(str(denom)):
        count += 1

print(f"There are {count} expansions from the first {N} where the numerator has more digits than the denominator.")
```
