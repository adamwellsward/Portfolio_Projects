```python
# concatenate function
def check_concats(a, b, ps):
    """Takes two prime numbers, a and b, and a list of primes, and checks if both
    concatenations of the numbers are still prime.
    """
    a_str = str(a)
    b_str = str(b)
    conc1 = int(a_str + b_str)
    conc2 = int(b_str + a_str)
    if max(conc1, conc2) > max(ps):
        raise Warning(f"not enough primes in ps with {a} and {b}")
    return (conc1 in ps) and (conc2 in ps)
```

```python
N = 1000000
p_list = primes_fast(N)
print(max(p_list))
p_set = set(p_list)
```

```python
for i, p1 in enumerate(p_list[1:]):
    group = [p1]
    for p2 in p_list[i+2:]:
        if all([check_concats(val, p2, p_set) for val in group]):
            group.append(p2)
        if len(group) == 5:
            print(f"Lowest sum with 5 primes is {sum(group)}")
            found = True
            print(group)
            break
    if found:
        break
```
